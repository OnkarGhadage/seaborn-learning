# Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
```

```python
sns.lineplot(data=df, x='year', y='passengers')
plt.show()
```

```python
# Remove confidence interval
sns.lineplot(data=df, x='year', y='passengers', errorbar=None)
plt.show()
```
