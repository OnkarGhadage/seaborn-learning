# Violin plot
It is box plot's smarter cousin. It also shows everything a box plot shows.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

## Basic plot

```python
sns.violinplot(data=df, x='day', y='total_bill')
plt.show()
```

## Adding `hue` - split by category

```python
sns.violinplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

### `split=True` - Mirror violins for two categories
When the `hue` has exactly 2 categories, we can split the violin in half

```python
sns.violinplot(data=df, x='day', y='total_bill', hue='sex', split=True)
plt.show()
```
