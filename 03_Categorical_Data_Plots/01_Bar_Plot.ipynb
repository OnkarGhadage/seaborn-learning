# Categorical Data Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Bar Plot
It shows average values of a numeric column for each categoru and automaticallu adds error bars.  
The bar height = mean of `total_bill`.  
The black line on top = confidence interval - Shows how much data varies

```python
# Basic use

sns.barplot(data=df,
           x='day',
           y='total_bill')
plt.show()
```

```python
## Adding hue - split by category

sns.barplot(data=df,
           x='day',
           y='total_bill',
           hue='sex',
           palette='Set2') # This uses color palette of colors
plt.show()
```

### Customizing

|parameter|info|ex|
|-|-|-|
|x|category col|'col_name'|
|y|Numeric col|'col_name'|
|hue|split by categoty|'col_name'|
|palette|color scheme|'Set2', 'viridis', 'coolwarm', 'Blues', 'husl'|
|capsize|error bar cap width|0.1, 0.2|
|edgecolor|bar border color|'black'|

```python
# EX

fig, axes = plt.subplots(1, 2)

sns.barplot(data=df, x='day', y='total_bill', capsize=0.1, ax=axes[0])
axes[0].set_title('1')

sns.barplot(data=df, x='sex', y='tip', ax=axes[1])
axes[1].set_title('2')

plt.suptitle('Title', fontsize=22)
plt.tight_layout()
plt.show()
```
