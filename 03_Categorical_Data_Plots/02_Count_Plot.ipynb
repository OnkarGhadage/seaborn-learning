# Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

## Basic usage
Counts how many rows belong to eachcategory and plots bars.

```python
sns.countplot(data=df, x='time')
plt.show()
sns.countplot(data=df, x='sex')
plt.show()
```

## `hue` - split by category

```python
sns.countplot(data=df, x='day', hue='sex')
plt.show()
```

## Horizontal

```python
sns.countplot(data=df, y='day', hue='sex') # Just use y instead of x
plt.show()
```
