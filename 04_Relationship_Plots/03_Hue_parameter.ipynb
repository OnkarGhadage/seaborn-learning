# Hue Parameter
hue is the single most powerfull parameter in seaborn.  
Works on both scatter and line plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', palette='Set2', alpha=0.7)
plt.show()
```

```python
df1 = sns.load_dataset('flights')
sns.lineplot(data=df1, x='year', y='passengers', hue='month', palette='tab10')
plt.show()
```

```python
sns.lineplot(data=df1, x='year', y='passengers', hue='month', palette='tab10')
plt.legend(title='Months', bbox_to_anchor=(1,1))      # Moves the legend outside the plot
plt.show()
```
