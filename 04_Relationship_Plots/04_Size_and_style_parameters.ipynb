# Size and style parameters

These add two more dimentions of information to plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

## Style - change dot/line shape by category

`hue` -   
`style` - change dot/line shape by category

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='smoker', palette='Set2', alpha=0.7)
plt.show()
```

## Size - change dot/line size by category

```python
sns.scatterplot(data=df, x='total_bill', y='tip', size='size')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', size='size', sizes=(30, 200))
plt.show()
```
