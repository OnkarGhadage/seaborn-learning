# Box plot
It shows the spread and distribution of data using 5 key statistics in one chart.  

**Whiskers** - Range of normal data (Q1 - 1.5 * IQR to Q3 + 1.5 * IQR)  
**Q3** - 75% of data is below this  
**Q2** - Median - 50% of data is below this  
**Q1*** - 25% of data is below this  
**outliers** - Dots beyond wiskers  

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.boxplot(data=df, x='day', y='total_bill')
plt.show()
```

## Add `hue` - split by category

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

### Customizing

|parameter|info|ex|
|-|-|-|
|x|category col|'col_name'|
|y|Numeric col||
|hue|split by category||
|palette|color scheme||
|width|box width||
|linewidth|border thickness||
|filterprops|outliers style||

### Horizontal box plot
just interchange x and y cols name

```python
sns.boxplot(data=df, x='total_bill', y='day')
plt.show()
```

## Spotting outliers - the main usecase

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 5))

for ax, col in zip(axes, ['total_bill', 'tip', 'size']):
    sns.boxplot(data=df, y=col,
               ax= ax,
               flierprops=dict(marker='o',
                                markerfacecolor='red',
                                markersize=7))

plt.tight_layout()
plt.show()
```
