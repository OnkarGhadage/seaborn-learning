# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.show()
```

## Custamization

|parameter|info|ex|
|-|-|-|
|s|dot size|80|
|alpha|transparancy||
|edgecolor|dot border||

```python
sns.scatterplot(data=df, x='total_bill', y='tip', s=40, alpha=0.4, edgecolor='black')
plt.show()
```
