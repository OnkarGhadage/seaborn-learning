# Strip Plot
It shows every single data point as a dot, groupes by category

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

## Basic use

```python
sns.stripplot(data=df, x='day', y='total_bill')
plt.show()
```

Every dot -> actual row in dataset

## `jitter` - spread dots so they dont overlap


```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.stripplot(data=df, x='day', y='total_bill', jitter=False, ax=axes[0])
axes[0].set_title('No jitter')
sns.stripplot(data=df, x='day', y='total_bill', jitter=True, ax=axes[1])
axes[1].set_title('With jitter')

plt.tight_layout()
plt.show()
```

## Adding `hue`

```python
sns.stripplot(data=df, x='day', y='total_bill', hue='sex', jitter=True)
plt.tight_layout()
plt.show()
```

### Custamizatiion

|parameter|info|ex|
|-|-|-|
|jitter||True/False|
|size|dot size|4, 5, 9|
|alpha|transparancy|0.0 to 1.0|
|edgecolor|dot border color|'color'|
|linewidth|border thickness|0.5, 1.0|
